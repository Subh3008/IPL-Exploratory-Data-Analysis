# cleaning the data

in the above data the players are segregated based on the following parameters 
 
 1.CAPPED-BATTERS,BOWLERS,ALLROUNDERS,WICKET-KEEPPER
 2.UNCAPPED-BATTERS,BOWLERS,ALLROUNDERS,WICKET-KEEPPER

THE UNCAPPED PLAYERS ARE PLAYING FOR THE FIRST TIME SO THERE IS NO RELEVANT DATA AVAILABLE ABOUT THEIR T20 PERFORMANCES



```python
#SEGREGATING DATA = CAPPED BATTERS
batters=df.loc[(df['Player_Type']=='Batter')]
batters_new=batters.loc[(batters['Capped']==1)]
capped_batters=batters_new[['Player Name','Team','Nationality','Matches_Played','Runs','Average','Strike_Rate']]
capped_batters.head()
```

```python
#cleaning the capped_batters data by making the null or NAN vcalues 0 using fillna()
#later checking if there is any null value present or not using isnull.sum()
capped_batters=capped_batters.fillna(0)
capped_batters.isnull().sum()
```

```python
#SEGRAGATING DATA=CAPPED ALLROUNDERS

allrounders=df.loc[(df['Player_Type'] =='Allrounder')]
allrounders_new=allrounders.loc[(allrounders['Capped']==1)]
capped_allrounders=allrounders_new[['Player Name','Matches_Played','Team','Nationality','Runs','Average','Strike_Rate','Wickets','Bowling_average','Economy','Bowling_Strike_Rate']]
capped_allrounders.head()
```

```python
#cleaning the capped_allrounderrs data by making the null or NAN vcalues 0 using fillna()
#later checking if there is any null value present or not using isnull.sum()

capped_allrounders=capped_allrounders.fillna(0)
capped_allrounders.isnull().sum()
```

```python
#SEGRAGATING DATA=CAPPED WICKET KEEPERS

WKs=df.loc[(df['Player_Type'] =='Keeper')]
WKs_new=WKs.loc[(WKs['Capped']==1)]
capped_WKs=WKs_new[['Player Name','Matches_Played','Team','Nationality','Runs','Average','Strike_Rate','Catches','Stumps','Run_outs']]
capped_WKs.head()
```

```python
# cleaning the capped_wicketkeepers data by making the null or NAN vcalues 0 using fillna()
#later checking if there is any null value present or not using isnull.sum()

capped_WKs=capped_WKs.fillna(0)
capped_WKs.isnull().sum()
```
